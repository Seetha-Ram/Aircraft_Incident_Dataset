```python
import plotly.graph_objs as go
```

```python
x = df['Incident_Category']
y0 = df['Aircraft_Phase']
fig = go.Figure()
fig.add_trace(go.Scatter(x=x, y=y0,
                    mode='lines+markers',
                    name='Incident_Category'))

fig.update_layout(legend_orientation="h",
                  legend=dict(x=0, y=1, traceorder="normal"),
                  title="Incident vs Aircraft phases",
                  xaxis_title="Incident_Category",
                  yaxis_title="Aircraft_Phase",
                  margin=dict(l=0, r=0, t=30, b=0))
fig.show()
```

```python
x = df['Aircaft_Damage_Type']
y0 = df['Incident_Category']
fig = go.Figure()
fig.add_trace(go.Scatter(x=x, y=y0,
                    mode='lines+markers',
                    name='Incident_Category'))

fig.update_layout(legend_orientation="h",
                  legend=dict(x=0, y=1, traceorder="normal"),
                  title="Incident_Category",
                  xaxis_title="Aircaft_Damage_Type",
                  yaxis_title="Incident_Category",
                  margin=dict(l=0, r=0, t=30, b=0))
fig.show()
```
